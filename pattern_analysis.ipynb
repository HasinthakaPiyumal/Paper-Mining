```python
import seaborn as sns,pandas as pd,matplotlib.pyplot as plt
```

```python
df = pd.read_csv("clustering/full_patterns_with_clusters.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(20,5))
sns.countplot(data=df,x="Cluster Number",palette="Set2")
plt.title("Distribution of Patterns across Clusters")
plt.show()
```

```python
plt.figure(figsize=(10,20))
sns.histplot(data=df, x="Cluster Number", hue="Category", multiple="stack", shrink=0.8)
plt.title("Pattern Categories per Cluster")
plt.show()
```

```python
grouped = df.groupby("Cluster Number")["Pattern Name"].apply(list).reset_index()

print(grouped)
```

```python
for _, row in grouped.iterrows():
    print(f"Cluster {row['Cluster Number']}:")
    for name in row["Pattern Name"]:
        print("  -", name)
    print()
```
